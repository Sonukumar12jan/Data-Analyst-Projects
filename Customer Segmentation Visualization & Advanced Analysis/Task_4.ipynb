## Tenure Groups

```python
import pandas as pd
```

### data read/load

```python
data = pd.read_csv("Telco_Customer_Churn_Dataset.csv")
```

```python
data["TotalCharges"] = pd.to_numeric(data["TotalCharges"], errors='coerce')
```

```python
data["TotalCharges"].fillna(data["TotalCharges"].mean(), inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(" ", "_")
```

```python
bins = [0, 12, 36, 100]
labels = ["0-12", "13-36", "37+"]

data["tenure_group"] = pd.cut(data["tenure"], bins=bins, labels=labels)
```

## Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
data["tenure_group"].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Customer Distribution by Tenure")
plt.show()
```

### Bar Chart (Avg Charges)

```python
import numpy as np
import seaborn as sns
```

```python
sns.barplot(x="tenure_group", y="monthlycharges", data=data)
plt.title("Avg Monthly Charges by Tenure Group")
plt.show()
```
